```python
#1. Concept Learning using Find-s Algorithm

data = [['Sunny','Warm','Yes'], ['Sunny','Cold','Yes'], ['Rainy','Warm','No']]
hypothesis = ['0','0']

for x in data:
    if x[-1] == 'Yes':
        for i in range(len(hypothesis)):
            if hypothesis[i] == '0':
                hypothesis[i] = x[i]
            elif hypothesis[i] != x[i]:
                hypothesis[i] = 'Yes'

print("Final Hypothesis:", hypothesis)
```

```python
#2. Candidate Elimination Algorithm

S = ['0','0']
G = ['?','?']

data = [['Sunny','Warm','Yes'], ['Rainy','Cold','No']]

for x in data:
    if x[-1]=='Yes':
        for i in range(2):
            if S[i]=='0': S[i]=x[i]
            elif S[i]!=x[i]: S[i]='?'
    else:
        for i in range(2):
            if x[i]!=S[i]: G[i]=S[i]

print("S:",S)
print("G:",G)
```

```python
#3. Inductive Bias in Learning Algorithm

hypotheses = ['h1','h2','h3']
errors = [2,0,1]

best = hypotheses[errors.index(min(errors))]
print("Selected Hypothesis:", best)
```

```python
#4. Decision Tree using ID3 Algorithm

import math

def entropy(p,n):
    if p==0 or n==0: return 0
    t=p+n
    return -p/t*math.log2(p/t)-n/t*math.log2(n/t)

print("Entropy:",entropy(9,5))
```

```python
#5. Heuristic Search for Hypothesis Space 

states = {'h1':3,'h2':1,'h3':2}
best = min(states, key=states.get)
print("Best Hypothesis:", best)
```
